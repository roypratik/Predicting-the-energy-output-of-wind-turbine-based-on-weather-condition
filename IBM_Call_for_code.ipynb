```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
my_filepath = "/home/roy/Desktop/T1.csv"
```

```python
my_data = pd.read_csv(my_filepath,index_col="Date/Time", parse_dates=True)
```

```python
my_data.head()
```

```python
list(my_data.columns)
```

```python
plt.figure(figsize=(20,15))
sns.lineplot(x=my_data['Wind Speed (m/s)'], y=my_data['LV ActivePower (kW)'])
```

```python
plt.figure(figsize=(20,15))
sns.lineplot(x=my_data['Wind Speed (m/s)'], y=my_data['Theoretical_Power_Curve (KWh)'])
```

```python
plt.figure(figsize=(20,15))
sns.lineplot(x=my_data['Wind Direction (°)'], y=my_data['Theoretical_Power_Curve (KWh)'])
```
